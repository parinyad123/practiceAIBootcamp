```
import tensorflowjs as tfjs

from tensorflow import keras
from tensorflow.keras.applications import MobileNet
from tensorflow.keras.layers import Dense,GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.applications.mobilenet import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator

input_your_code = None
```

```
'''Define a base model from MobileNet without the network head
Transfer weights from MobileNet training from ImageNet dataset.
'''
base_model = MobileNet(
    weights='imagenet',
    input_shape=(224, 224, 3),
    include_top=False
)
```

```
x = base_model.output
x = input_your_code

# Add dense layers so that the model can learn more complex functions and classify for better results.
# The first Dense layer here has 1024 outputs with ReLU activation
x = input_your_code

# Dense layer 2 with 1024 outputs and ReLU activation
x = input_your_code

# Dense layer 3 with 512 outputs and ReLU activation
x = input_your_code

# Final Dense layer with softmax activation, the number of class if 4
predictions = input_your_code
```

```
# inputs is training set (number of training set, 224, 224, 3)
# outputs is the layers defined previously.
model=Model(
    inputs=base_model.input,
    outputs=predictions
)
```

```
# List of all layers.
# You can select which layers you want to freeze the weights.
for i, layer in enumerate(model.layers):
  print(i, layer.name)
```

```
# Freeze the first 20 layers
freeze_before_layer = 20
for i, layer in enumerate(model.layers[:freeze_before_layer]):
    layer.trainable=False
    print(i, layer.name)
```

```
# Update weights for the rest layers.
for i, layer in enumerate(model.layers[20:]):
    layer.trainable=True
    print(i + freeze_before_layer, layer.name)
```

```
model.summary()
```

```
# See more parameters: https://keras.io/preprocessing/image/
# Adjust data augmentation here:
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=2,
    shear_range=0.1,
    zoom_range=0.2
)
```

```
# Generate training set from pacman_dataset_v2.
# Adjust batch size here depending on number of samples and GPU memory.
train_generator=train_datagen.flow_from_directory(
    'dataset/pacman_dataset_v2/train',
    target_size=(224, 224),
    color_mode='rgb',
    batch_size=8,
    class_mode='categorical'
)
```

```
# Generate validation set from pacman_dataset_v2
validation_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
validation_generator = validation_datagen.flow_from_directory(
    directory='dataset/pacman_dataset_v2/val',
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=1,
    class_mode="categorical"
)
```

```
'''Create TensorFlow model graph.
Adam is similar to stochastic gradient descent but it is automatically adjust
learning rate.
'''
model.compile(
    optimizer='Adam',
    loss=keras.losses.categorical_crossentropy,
    metrics=['accuracy']
)
```

```
# Step size depends on your batch size, 1 epoch should run through all samples.
step_size_train = train_generator.n // train_generator.batch_size
step_size_validation = validation_generator.n // validation_generator.batch_size

# Run training and validation
model.fit_generator(
    generator=train_generator,
    steps_per_epoch=step_size_train,
    epochs=3,
    validation_data=validation_generator,
    validation_steps=step_size_validation
)
```

```
# Convert a model to TensorFlow.js JSON format
tfjs.converters.save_keras_model(model, 'models/')
```
